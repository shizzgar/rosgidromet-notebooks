# Обработка и визуализация данных о погоде

## Подготовка
Для начала импортируем необходимые библиотеки. 

Сначала импортируем **pandas**
>**pandas** — программная библиотека на языке Python для обработки и анализа данных. Работа pandas с данными строится поверх библиотеки NumPy, являющейся инструментом более низкого уровня. Предоставляет специальные структуры данных и операции для манипулирования числовыми таблицами и временны́ми рядами. 

Следом **NumPy**
>**NumPy** — библиотека с открытым исходным кодом для языка программирования Python. Возможности: поддержка многомерных массивов; поддержка высокоуровневых математических функций, предназначенных для работы с многомерными массивами.

и наконец **matplotlib**
>**matplotlib** — библиотека на языке программирования Python для визуализации данных двумерной графикой. Получаемые изображения могут быть использованы в качестве иллюстраций в публикациях. Matplotlib написан и поддерживался в основном Джоном Хантером и распространяется на условиях BSD-подобной лицензии.

Теперь прочитаем файл в переменную `df`, сокращение от dataFrame, с помощью функции `read_csv`, где в параметрах укажем названием файла `"data.csv"` и подскажем pandas'y, что в первой колонке у нас дни за счет `dayfirst=True`. 
Теперь можно взглянуть на наши данные. 
Посмотрим первые `5` строк нашей таблици с помощью `head(5)`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("data.csv", dayfirst=True)
df.head(5)               
```

## График одного года
В переменной `year` создаем новый DataFrame, на основе средних данных за июнь 2000-го года `['2000']`, заранее зачищая первую строку с помощью `[1:]`
Так же при создании добавим, что данные необходимо преобразовать в тип _значений с плавающей точкой_.
И с помощью функции `plot()` создаим график на основе данных записанных в переменную `year`

```python
year = pd.DataFrame(data=df['2000'][1:], dtype=np.float)
year.plot()
```

## Визуализация данных за 20 лет


```python
year = lambda df, x : df[str(1998+x)][1:]
years = []
for x in range(21):
    years.append(year(df,x))
twenty = pd.DataFrame(data=years, dtype=np.float)
twenty.T.plot(figsize=(20, 10), grid=True)
```
